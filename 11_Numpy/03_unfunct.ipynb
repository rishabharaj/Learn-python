```python
# --- 1. ufunc Intro ---
import numpy as np
print("Type of np.add:", type(np.add))
```

```python
# --- 2. ufunc Create Function ---
import numpy as np
def myadd(x, y):
  return x+y
myadd = np.frompyfunc(myadd, 2, 1)
print("Custom ufunc Add:", myadd([1, 2, 3, 4], [5, 6, 7, 8]))
```

```python
# --- 3. ufunc Simple Arithmetic ---
import numpy as np
arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([3, 5, 6, 8, 2, 33])
print("Add:", np.add(arr1, arr2))
print("Subtract:", np.subtract(arr1, arr2))
print("Power:", np.power(arr1, arr2))
```

```python
# --- 4. ufunc Rounding Decimals ---
import numpy as np
print("Truncate:", np.trunc([-3.1666, 3.6667]))
print("Around:", np.around(3.1666, 2))
```

```python
# --- 5. ufunc Logs ---
import numpy as np
arr = np.arange(1, 10)
print("Log2:", np.log2(arr))
print("Log10:", np.log10(arr))
```

```python
# --- 6. ufunc Summations ---
import numpy as np
arr1 = np.array([1, 2, 3])
arr2 = np.array([1, 2, 3])
print("Sum:", np.sum([arr1, arr2]))
print("Sum over axis:", np.sum([arr1, arr2], axis=1))
print("Cumsum:", np.cumsum(arr1))
```

```python
# --- 7. ufunc Products ---
import numpy as np
arr = np.array([1, 2, 3, 4])
print("Product:", np.prod(arr))
print("Cumprod:", np.cumprod(arr))
```

```python
# --- 8. ufunc Differences ---
import numpy as np
arr = np.array([10, 15, 25, 5])
print("Difference:", np.diff(arr))
```

```python
# --- 9. ufunc Finding LCM ---
import numpy as np
arr = np.array([3, 6, 9])
print("LCM of array:", np.lcm.reduce(arr))
```

```python
# --- 10. ufunc Finding GCD ---
import numpy as np
arr = np.array([20, 8, 32, 36, 16])
print("GCD of array:", np.gcd.reduce(arr))
```

```python
# --- 11. ufunc Trigonometric ---
import numpy as np
print("Sin(pi/2):", np.sin(np.pi/2))
arr = np.array([90, 180, 270, 360])
print("Degrees to Radians:", np.deg2rad(arr))
```

```python
# --- 12. ufunc Hyperbolic ---
import numpy as np
print("Sinh(pi/2):", np.sinh(np.pi/2))
arr = np.array([0.1, 0.2, 0.5])
print("Arcsinh:", np.arcsinh(arr))
```

```python
# --- 13. ufunc Set Operations ---
import numpy as np
set1 = np.array([1, 2, 3, 4])
set2 = np.array([3, 4, 5, 6])
print("Unique:", np.unique([1, 1, 1, 2, 3, 4, 5, 5, 6, 7]))
print("Union:", np.union1d(set1, set2))
print("Intersection:", np.intersect1d(set1, set2, assume_unique=True))
print("Difference:", np.setdiff1d(set1, set2, assume_unique=True))
```
