```python
from numpy import random
```

```python
# --- 1. Random Intro ---
x = random.randint(100)
x_float = random.rand()
arr_rand = random.randint(100, size=(5))
print("Random Integer:", x)
```

```python
# --- 2. Data Distribution ---
x_dist = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(5))
print("Distribution Array:", x_dist)
```

```python
# --- 3. Random Permutation ---
import numpy as np
arr_perm = np.array([1, 2, 3, 4, 5])
random.shuffle(arr_perm)
print("Shuffled:", arr_perm)
```

```python
# --- 5. Normal Distribution ---
x_norm = random.normal(loc=1, scale=2, size=(2, 3))
print("Normal Dist:\n", x_norm)
```

```python
# --- 6. Binomial Distribution ---
x_bino = random.binomial(n=10, p=0.5, size=10)
print("Binomial Dist:", x_bino)
```

```python
# --- 7. Poisson Distribution ---
x_pois = random.poisson(lam=2, size=10)
print("Poisson Dist:", x_pois)
```

```python
# --- 8. Uniform Distribution ---
x_uni = random.uniform(size=(2, 3))
print("Uniform Dist:\n", x_uni)
```

```python
# --- 9. Logistic Distribution ---
x_logi = random.logistic(loc=1, scale=2, size=(2, 3))
print("Logistic Dist:\n", x_logi)
```

```python
# --- 10. Multinomial Distribution ---
x_multi = random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
print("Multinomial Dist:", x_multi)
```

```python
# --- 11. Exponential Distribution ---
x_exp = random.exponential(scale=2, size=(2, 3))
print("Exponential Dist:\n", x_exp)
```

```python
# --- 12. Chi Square Distribution ---
x_chi = random.chisquare(df=2, size=(2, 3))
print("Chi Square Dist:\n", x_chi)
```

```python
# --- 13. Rayleigh Distribution ---
x_ray = random.rayleigh(scale=2, size=(2, 3))
print("Rayleigh Dist:\n", x_ray)
```

```python
# --- 14. Pareto Distribution ---
x_par = random.pareto(a=2, size=(2, 3))
print("Pareto Dist:\n", x_par)
```

```python
# --- 15. Zipf Distribution ---
x_zipf = random.zipf(a=2, size=(2, 3))
print("Zipf Dist:\n", x_zipf)
```
